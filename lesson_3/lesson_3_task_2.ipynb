```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-5, 5, 21)
y = 3*x+1
y2 = (-1/3)*x+1
plt.plot(x,y)
plt.plot(x,y2)
plt.xlabel("x")
plt.ylabel("y")
```

```python
plt.figure(figsize = (5, 5))
x = np.linspace(-15, 15)
y = 3*x+1
y2 = (-1/3)*x+1
plt.plot(x,y)
plt.plot(x,y2)
plt.xlabel("x")
plt.ylabel("y")
plt.ylim(-15, 15)
plt.xlim(-15, 15)
plt.show()
```

Прямые не кажутся перпендикулярными из-за системы координат
