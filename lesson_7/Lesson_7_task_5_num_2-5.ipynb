```python
# 5.4. скалярное произведение двух векторов:(1, 5)  и  (2, 8)
A = np.array([1,5])
B = np.array([2,8])
print("Скалярное произведение A и B:", np.dot(A,B))

x,y = np.array([0, 0]), np.array([0, 0])
u, v = np.array([A[0], B[0]]), np.array([A[1], B[1]])
plt.quiver(x,y,u,v, angles='xy', scale_units='xy', scale=1)
plt.xlim(-10, 10)
plt.ylim(-10,10)
plt.grid(True)
plt.show()
```

```python
# 5.5 смешанное произведение трех векторов:(1, 5, 0), (2, 8, 7) и (7, 1.5, 3)

A = np.array([1,5,0])
B = np.array([2,8,7])
C = np.array([7,1.5,3])

# способ 1: необходимо найти определитель системы, составленной из координат векторов.
D = np.array([A,B,C])
print("матрица D",D)
print("Определитель матрицы D:", np.linalg.det(D))

# способ 2: найти векторное произведение A и B, умножить скалярно на вектор C
m1 = np.cross(A,B)
print("Векторное произведение A и B:",m1)
print("Векторное произведение A и B, умноженное скалярно на C:",np.dot(m1,C))
```
