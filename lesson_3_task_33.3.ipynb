```python
def polar2cart(r,phi):
    x = r*math.cos(phi)
    y = r*math.sin(phi)
    return x,y

polar2cart(2,45)
```

```python
phi = np.arange(0., 2., 1./180.)*np.pi 

plt.polar(phi, [10]*len(phi)) 

plt.show()
```
